```python
from pyforest import *
```

```python
df1 = pd.read_csv('Plant_1_Generation_Data.csv')
df2 = pd.read_csv('Plant_1_Weather_Sensor_Data.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1 = df1.drop(['PLANT_ID', 'SOURCE_KEY'], axis =1 )
df2 = df2.drop(['PLANT_ID', 'SOURCE_KEY'], axis = 1)
```

```python
df1['DATE_TIME'] = pd.to_datetime(df1['DATE_TIME'])
df2['DATE_TIME'] = pd.to_datetime(df2['DATE_TIME'])
```

```python
df1['HOUR'] = df1['DATE_TIME'].dt.hour
df1['DAY'] = df1['DATE_TIME'].dt.dayofyear
df1['MONTH'] = df1['DATE_TIME'].dt.month

df1.head()
```

```python

df1 = df1.set_index('DATE_TIME')
df1_i5mins = df1.resample('15T').mean()

df1_15mins = df1_i5mins.reset_index()
```

```python
final_df = pd.merge(df1_15mins, df2, on = 'DATE_TIME')
final_df
```

```python
final_df.info()
```

```python
df = final_df.drop(['DAILY_YIELD', 'TOTAL_YIELD','MONTH', 'DAY', 'DATE_TIME'], axis = 1)
df.head()
```

```python
df.HOUR.value_counts()
```

```python
df.to_csv('final_solar_dataset.csv', index = False)
```
